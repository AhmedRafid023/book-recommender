```python
import kagglehub
from seaborn import heatmap
from sympy import transpose

# Download latest version
path = kagglehub.dataset_download("dylanjcastillo/7k-books-with-metadata")

print("Path to dataset files:", path)
```

```python
import pandas as pd
```

```python
books = pd.read_csv(f"{path}/books.csv")
```

```python
books
```

**Create a seaborn heatmap of missing values**

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
ax = plt.axes()
sns.heatmap(books.isna().transpose(), cbar=False, ax=ax)

plt.xlabel("Columns")
plt.ylabel("Missing Values")
plt.show()
```

**Books where description is missing put a 1 in that column or put a 0 in that column**

```python
import numpy as np

books["missing_description"] = np.where(books["description"].isna(), 1, 0)
books["age_of_book"] = 2025 - books["published_year"]
```

```python
column_of_interest = ["num_pages", "age_of_book", "missing_description", "average_rating"]

correlation_matrix = books[column_of_interest].corr(method="spearman")

sns.set_theme(style="white")
plt.figure(figsize=(8, 6))
heatmap = sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", cbar_kws={"label": "Spearman Correlation"})

heatmap.set_title("Correlation Heatmap")
plt.show()

```

**my data is 0 and 1 so i use the spearman method. if the  data is  continuous we will use pearson method**

**Here we want to findout missing description correlation with other variable.We can see that there is no value close to 1 as a result we can say there is no strong correlation**

**We are going to delete the books with missing description. At first lets find out how many books will this cost us**

```python
book_without_missing = books[
      (~books["description"].isna()) &
      (~books["average_rating"].isna()) &
      (~books["published_year"].isna()) &
      (~books["num_pages"].isna())
]

book_without_missing
```

**SORT BY CATEGORY**

```python
book_without_missing["categories"].value_counts().sort_values(ascending=False)
```

**Lets now refine the description categories. The description needs to be meaningful. So we will need the description with atleast more than 25 words**

```python
book_without_missing.loc[:, "words_in_description"] = book_without_missing["description"].str.split().str.len()

book_without_missing
```

```python
book_without_missing.loc[book_without_missing["words_in_description"].between(1 , 4), "description"]
```

```python
book_without_missing.loc[book_without_missing["words_in_description"].between(5 , 14), "description"]
```

```python
book_without_missing.loc[book_without_missing["words_in_description"].between(15 , 24), "description"]
```

```python
book_without_missing_25_words = book_without_missing[book_without_missing["words_in_description"] >= 25]
```

```python
book_without_missing_25_words
```

**Lets fix the subtitle. we can create a field title and subtitle and we will use subtitle here as the data when subtitle is missing , we will just use the title as the data**

```python
book_without_missing_25_words.loc[:, "title_and_subtitle"] = np.where(
    book_without_missing_25_words["subtitle"].isna(),
    book_without_missing_25_words["title"],
    book_without_missing_25_words[["title", "subtitle"]].astype(str).agg(": ".join, axis=1)
)
```

```python
book_without_missing_25_words
```

```python
book_without_missing_25_words.loc[:, "tagged_description"] = book_without_missing_25_words[["isbn13", "description"]].astype(str).agg(" ".join, axis=1)
```

```python
book_without_missing_25_words
```

**Drop the unnecessary column and make the clean dataset**

```python
book_without_missing_25_words.drop(["subtitle", "missing_description", "age_of_book", "words_in_description"], axis=1).to_csv("books_cleaned.csv", index=False)
```
